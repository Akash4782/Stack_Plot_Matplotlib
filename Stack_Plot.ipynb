Stack Plot

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
area = [1,2,3,4,5]

plt.stackplot(x,area)

plt.show()
```

```python
x = [1,2,3,4,5]
area = [2,3,2,5,4]

plt.stackplot(x,area)

plt.show()
```

```python
x = [1,2,3,4,5]
area = [2,3,2,5,4]
area1 = [2,3,4,5,6]
area2 = [1,3,4,5,2]

plt.stackplot(x,area,area1 , area2)

plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,4,5,2]

l =["area1","area2","area3"]
plt.stackplot(x,area1,area2 , area3 , labels = l)
plt.legend(loc = 2)

plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,4,5,2]

l =["area1","area2","area3"]

c = ["r","g","y"]
plt.stackplot(x,area1,area2 , area3 , labels = l , colors = c)
plt.legend(loc = 2)

plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,4,5,2]

l =["area1","area2","area3"]

c = ["r","g","m"]
plt.stackplot(x,area1,area2 , area3 , labels = l , colors = c , baseline = "zero")
plt.legend(loc = 2)

plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,4,5,2]

l =["area1","area2","area3"]

c = ["r","g","m"]
plt.stackplot(x,area1,area2 , area3 , labels = l , colors = c , baseline = "sym")
plt.legend(loc = 2)

plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,4,5,2]

l =["area1","area2","area3"]

c = ["r","g","m"]
plt.stackplot(x,area1,area2 , area3 , labels = l , colors = c , baseline = "wiggle")
plt.legend(loc = 2)

plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,4,5,2]

l =["area1","area2","area3"]

c = ["r","g","m"]
plt.stackplot(x,area1,area2 , area3 , labels = l , colors = c)
plt.legend(loc = 2)
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,4,5,2]

l =["area1","area2","area3"]

c = ["r","g","m"]
plt.stackplot(x,area1,area2 , area3 , labels = l , colors = c)
plt.legend(loc = 2)
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```
